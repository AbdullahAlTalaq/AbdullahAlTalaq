```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
df = pd.read_csv('/content/Pancreatic Cancer (cleaned).csv')
```

```python
# Histogram for age
plt.figure(figsize=(8,6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# Bar plot for diagnosis
#differentiating between 3 (pancreatic cancer) versus 2 (non-cancerous pancreas condition) and 1 (healthy).
plt.figure(figsize=(8,6))
sns.countplot(x='diagnosis',data=df)
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis')
plt.show()
```

```python
# Correlation matrix and heatmap
#correlation heatmap to explore relationships between numerical features.
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairwise scatter plots
custom_palette = {1: 'blue', 2: 'green', 3: 'red'}
sns.pairplot(df[['age','diagnosis','creatinine', 'LYVE1', 'REG1B','REG1A', 'TFF1']],diag_kind='kde', hue='diagnosis', palette=custom_palette)
plt.suptitle('Pairwise Scatter Plots')
plt.show()
```

```python
# Violin plot for age by diagnosis
#plots to visualize the distribution of a numerical feature across different diagnosis categories, providing more information than box plots.
plt.figure(figsize=(10,8))
sns.violinplot(x='diagnosis',y='age',data=df)
plt.title('Age Distribution by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Age')
plt.show()
```

```python
# Violin plot for creatinine by sex
plt.figure(figsize=(10, 8))
sns.violinplot(x='sex', y='diagnosis', data=df)
plt.title('Creatinine Levels by Sex')
plt.xlabel('Sex')
plt.ylabel('Creatinine')
plt.show()
```

#the relationship between REG1A and REG1B.


```python
sns.scatterplot(x='REG1B', y='REG1A', data=df, hue='diagnosis', palette=custom_palette)
plt.title('Scatter Plot of REG1B vs. REG1A')
plt.xlabel('REG1B')
plt.ylabel('REG1A')
plt.show()
```

# the relationship between age and creatinine.

```python
sns.scatterplot(x='age', y='creatinine', data=df, hue='diagnosis', palette=custom_palette)
plt.title('Scatter Plot of Age vs. Creatinine')
plt.xlabel('Age')
plt.ylabel('Creatinine')
plt.show()
```

# the relationship between age and plasma_CA19_9 levels.

```python
sns.scatterplot(x='age', y='plasma_CA19_9', data=df, hue='diagnosis', palette=custom_palette)
plt.title('Scatter Plot of Age vs. Plasma_CA19_9')
plt.xlabel('Age')
plt.ylabel('Plasma_CA19_9')
plt.show()
```
